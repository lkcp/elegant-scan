```python
#reduce qm from 10000 to 1000

import numpy as np

p = open('Cy.txt')
f = p.readlines()
output = []
n = np.linspace(1,9999,1000)
for i in range(len(n)):
    line = int(n[i])
    output.append(f[line])

newflie = open('Cy1.txt','w')
newflie.writelines(output)
newflie.close()
    
```
